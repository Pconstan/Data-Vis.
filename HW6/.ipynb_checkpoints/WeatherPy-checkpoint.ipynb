```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import csv
import pprint
```

```python
from api_keys import api_key
```

```python
from citipy import citipy
```

```python
output_data_file = "output_data/cities.csv"
```

```python
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
#List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
w_compiled = []
num = 1
set_ = 1
url = "http://api.openweathermap.org/data/2.5/weather?"
print(f'''
          Beginning Data Extraction
          --------------------------''')
for city in cities:
    query_url = url + "appid=" + api_key + "&q=" + city
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()
    w_compiled.append(weather_json)
    print(f'''
          Processing Record {num} of Set {set_} | {city}
          ''')
    num = num+1
    if num > 50:
        set_ = set_ +1
        num = 1
print('''----------------
        Data Extraction Complete
        ''')
```

```python
# Converting to df
w_df = pd.DataFrame(w_compiled)
# dropping n/a's
w_df = w_df.dropna(thresh=5)
w_df = w_df.reset_index()
```

```python
#isolating only the parts we need for the frame
wt = []
wh = []
wc = []
wcl = []
wlat = []
wlon = []
wwin = []

for i in range(0,len(w_df)):
    t = (w_df.main[i]['temp_max']-273.15)*(9/5)+32
    h = w_df.main[i]['humidity']
    c = w_df.sys[i]['country']
    cl = w_df.clouds[i]['all']
    lat = w_df.coord[i]['lat']
    lon = w_df.coord[i]['lon']
    win = w_df.wind[i]['speed']
    wt.append(t)
    wh.append(h)
    wc.append(c)
    wcl.append(cl)
    wlat.append(lat)
    wlon.append(lon)
    wwin.append(win)
    
```

```python
#new column names
w_dict = {"City":w_df.name, "Cloudiness":wcl, "Country":wc, 
          "Date": w_df.dt, "Humidity":wh, "Lat":wlat, "Lon":wlon,
          "Max Temp": wt, "Wind Speed": wwin}
```

```python
#our final dataframe
w_c_df = pd.DataFrame(w_dict)
w_c_df.to_csv('weather_latitude.csv')
```

```python
#Temperature (F) vs. Latitude
y = w_c_df["Max Temp"]
x = w_c_df["Lat"]
plt.scatter(x,y)
plt.title("City Latitude vs. Max Temp. (1/19/19)")
plt.xlabel('Latitude')
plt.ylabel('Max Temp. (F)')
plt.grid(True)
plt.savefig('city_lat_max_temp.png', bbox_inches='tight')
plt.show()
```

```python
#Humidity (%) vs. Latitude
y = w_c_df["Humidity"]
x = w_c_df["Lat"]
plt.scatter(x,y)
plt.title("City Latitude vs. Humidity (1/19/19)")
plt.xlabel('Latitude')
plt.ylabel('Humidity')
plt.grid(True)
plt.savefig('city_lat_humidity.png', bbox_inches='tight')
plt.show()
```

```python
#Cloudiness (%) vs. Latitude
y = w_c_df["Cloudiness"]
x = w_c_df["Lat"]
plt.scatter(x,y)
plt.title("City Latitude vs. Cloudiness (1/19/19)")
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.grid(True)
plt.savefig('city_lat_cloudiness.png', bbox_inches='tight')
plt.show()
```

```python
#Wind Speed (mph) vs. Latitude
y = w_c_df["Wind Speed"]
x = w_c_df["Lat"]
plt.scatter(x,y)
plt.title("City Latitude vs. Wind Speed (1/19/19)")
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.grid(True)
plt.savefig('city_lat_wind_speed.png', bbox_inches='tight')
plt.show()
```
