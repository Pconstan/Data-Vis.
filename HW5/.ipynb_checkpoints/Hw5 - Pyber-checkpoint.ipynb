```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"


# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

```

```python
# Combine the data into a single dataset
joined_df = pd.merge(city_df, ride_df, on='city')
# Display the data table for preview
joined_df.head()
```

```python
# Obtain the x and y coordinates for each of the three city types
num_rides_per_city = joined_df.city.value_counts()
x_ax_len = num_rides_per_city.max()

avg_fares_by_city_df = joined_df.groupby(['type','city']).mean()
y_ax_len = avg_fares_by_city_df.fare.max()

#the number of unique cities by type
no_types = joined_df.groupby(['type']).nunique()
rural_len = no_types.iloc[0,0]
suburb_len = no_types.iloc[1,0]
urb_len = no_types.iloc[2,0]

num_rides_per_city_ordered = joined_df.groupby(['type','city']).nunique()

x_vals_list_r = num_rides_per_city_ordered.iloc[0:rural_len,4]
x_vals_list_s = num_rides_per_city_ordered.iloc[rural_len:suburb_len+rural_len,4]
x_vals_list_u = num_rides_per_city_ordered.iloc[suburb_len+rural_len:rural_len+urb_len+suburb_len,4]


y_vals_list_r = avg_fares_by_city_df.iloc[0:rural_len,1]
y_vals_list_s = avg_fares_by_city_df.iloc[rural_len:suburb_len+rural_len,1]
y_vals_list_u = avg_fares_by_city_df.iloc[suburb_len+rural_len:rural_len+urb_len+suburb_len,1]









 
```

```python
#determining sizes of bubbles
driver_1 = city_df.sort_values(['type','city'])
driver_r = driver_1.iloc[0:rural_len,1]
driver_s = driver_1.iloc[rural_len:suburb_len+rural_len,1]
driver_u = driver_1.iloc[suburb_len+rural_len:rural_len+urb_len+suburb_len,1]
area_r = driver_r**1.5
area_s = driver_s**1.5
area_u = driver_u**1.5
```

```python
# Build the scatter plots for each city types
rural = plt.scatter(x_vals_list_r,y_vals_list_r, s=area_r, c = 'Gold', alpha =.8, edgecolor = 'black')
suburban = plt.scatter(x_vals_list_s,y_vals_list_s, s=area_s, c = 'lightblue', alpha =.8, edgecolor = 'black')
urban = plt.scatter(x_vals_list_u,y_vals_list_u, s=area_u, c = 'Coral', alpha =.8, edgecolor = 'black')
# Create a legend
plt.legend(handles=[rural, suburban, urban], labels = ['Rural', 'Suburban', 'Urban'], title = 'City Types', markerscale = .7, loc="best")
# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
# Incorporate a text label regarding circle size
plt.text(1.02, 0.5, "Note: Circle size correlates with driver count per city", {'color': 'black', 'fontsize': 8},
         horizontalalignment='left',
         verticalalignment='center',
         clip_on=False,
         transform=plt.gca().transAxes)


# Save Figure
plt.savefig("Pyber.png")
plt.show()
```

```python
# Show plot
plt.show() 
```

```python
#Total Fares by City Type
#------------------------#
# Calculate Type Percents
tot_fares_by_type = joined_df.groupby(['type'])['fare'].sum()
#labels
labs = 'Rural','Suburban','Urban'
sizes = [tot_fares_by_type[0], tot_fares_by_type[1], tot_fares_by_type[2]]
colors = ['Gold','lightblue','coral']
explode = (0, 0, .1)

# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labs, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('% of Total Fares by City Type')
# Save Figure
plt.savefig("PyberPie_1.png")
plt.show()
```

```python
#Total Rides by City Type
#--------------------------#

tot_rides_by_type = joined_df.groupby(['type'])['fare'].count()
#labels
labs = 'Rural','Suburban','Urban'
sizes = [tot_rides_by_type[0], tot_rides_by_type[1], tot_rides_by_type[2]]
colors = ['Gold','lightblue','coral']
explode = (0, 0, .1)

# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labs, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('% of Total Rides by City Type')
# Save Figure
plt.savefig("PyberPie_2.png")
plt.show()
```

```python
#Total Drivers by City Type
#-------------------------#

driver_df = city_df.sort_values(['type','city'])
num_driver_r = driver_df.iloc[0:rural_len,1]
num_driver_s = driver_df.iloc[rural_len:suburb_len+rural_len,1]
num_driver_u = driver_df.iloc[suburb_len+rural_len:rural_len+urb_len+suburb_len,1]

tot_drivers_by_type = city_df.groupby(['type'])['driver_count'].count()
#labels
labs = 'Rural','Suburban','Urban'
sizes = [num_driver_r.sum(), num_driver_s.sum(), num_driver_u.sum()]
colors = ['Gold','lightblue','coral']
explode = (0, 0, .1)

# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labs, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('% of Total Drivers by City Type')
# Save Figure
plt.savefig("PyberPie_3.png")
plt.show()
```
